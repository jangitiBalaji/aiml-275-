```python
plt.figure(figsize=(6,3))
plt.title("Box plot for Daily Sales")
plt.boxplot(data1["daily"], vert = False)
plt.show()
```

```python
sns.histplot(data1['daily'], kde = True,stat='density',)
plt.show()
```

observations
1. there are no missing values
2. the daily column values apperas to be might-skewed
3. the sunday column values also appear to be right-skewed
4. there are two outliers in both daily column and aslo in sunday column as observed from the boxplots

```python
x = data1["daily"]
y = data1["sunday"]
plt.scatter(data1["daily"], data1["sunday"])
plt.xlim(0, max(x) + 100)
plt.ylim(0, max(y) + 100)
plt.show()
```
